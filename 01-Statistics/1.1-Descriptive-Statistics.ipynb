#### ğŸ“Š **Measure of Dispersion**

<div class="small-text">
When we calculate <b>Mean</b>, <b>Median</b>, or <b>Mode</b>, we know the <b>Center</b> of the data where most values lie.  
But we donâ€™t know how <b>spread out</b> the data is. Thatâ€™s where the <b>Measure of dispersion</b> comes in.
</div>

---

#### ğŸ§  **Simple Definition**
<div class="small-text">
A measure of dispersion tells us <b>how much the data values vary or spread out</b> around the mean (or any central value).
</div>

---

#### ğŸ¯ **Why Itâ€™s Important (Especially in ML)**
<div class="small-text">

- ğŸ“ˆ Understand **data consistency** â€” whether values are tightly packed or widely spread  
- ğŸ§® Many ML models (like regression, clustering, anomaly detection) assume certain **variance** or **normal distribution**  
- ğŸš¨ Helps **detect outliers** or high variability which can affect model accuracy  

</div>

#### ğŸ¯ <b>Sampling Techniques (Overview)</b>
<div class="small-text">
<b>Sampling</b> = choosing a subset from the population for analysis.  
Main goal â†’ ensure the <b>sample represents the population fairly</b>.
</div>

---

#### âš™ï¸ <b>Common Sampling Techniques</b>
<div class="small-text">

- ğŸ² <b>Simple Random Sampling:</b> Every item has an equal chance of being chosen.  
- ğŸ”¢ <b>Systematic Sampling:</b> Select every k<sup>th</sup> item after a random start.  
- ğŸ§© <b>Stratified Sampling:</b> Divide population into groups (strata) and sample proportionally from each.  
- ğŸŒ <b>Cluster Sampling:</b> Divide population into clusters, then randomly select entire clusters.

</div>

<div class="example-box">
<b>Example:</b>  
If your population is 10,000 students:  
- Pick 100 random students â†’ <b>Simple Random</b>  
- Pick every 50th student â†’ <b>Systematic</b>  
- Pick 25 students from each department â†’ <b>Stratified</b>  
- Pick 2 random departments and include all their students â†’ <b>Cluster</b>
</div>


```python
import pandas as pd 
import numpy as np
```

```python
""" 
ğŸ§® 1. Simple Random Sampling (SRS)

Definition:
Every item in the population has an equal chance of being selected.

Use Case:
When your data is uniform and thereâ€™s no subgroup difference (like students of same class).

"""
#population of 1000 values say marks
population = np.arange(1,1001)

#Take random sample of 10
sample = np.random.choice(population, size = 10, replace=False)

print("Sample: ", sample)
print("Sample mean: ", np.mean(sample))
print("Population mean: ", np.mean(population))

#The sample mean will be close to the population mean if sampling is fair.

```

```python
"""
Stratified Sampling

ğŸ¯ Definition:
Stratified sampling means dividing the population into groups (strata) based on some important characteristic then taking random samples from each group proportionally.

ğŸ’¡ Why we use it:
To make sure all key groups are represented fairly in the sample.
ğŸ‘‰ Useful when data has clear subgroups with different characteristics.

ğŸ§  Example:
You want to study average marks of students in your college, and your population has:

60% BCA students
30% BBA students
10% MCA students

Instead of taking a random sample (which might ignore some MCA students),
you take proportional samples:

60 from BCA
30 from BBA
10 from MCA

Thatâ€™s stratified sampling
"""

from sklearn.model_selection import train_test_split

data = pd.DataFrame({
    'student_id': range(1, 101),
    'course': ['BCA']*60 + ['BBA']*30 + ['MCA']*10
})

# Stratified sample (preserves course proportion)
train, sample = train_test_split(data, test_size=0.2, stratify=data['course'])

print(sample['course'].value_counts())
```

```python
""" 
Systematic Sampling

ğŸ¯ Definition:
Systematic sampling means selecting every káµ—Ê° item from the population after choosing a random starting point.

ğŸ’¡ Why we use it:
Simple and fast (no random selection each time)
Works well when data is evenly distributed
Not good if thereâ€™s a repeating pattern â€” it can introduce bias

ğŸ§  Example:
You have 1000 students, and you need a sample of 100 â†’
â†’ pick every 10th student after a random start.

"""

population = np.arange(1, 1001)  # 1000 students
k = 10  # sampling interval
start = np.random.randint(0, k)  # random start between 0â€“9

sample = population[start::k]
print("Sample size:", len(sample))
print("First 10 samples:", sample[:10])
```

```python
""" 
Convenience Sampling 

ğŸ¯ Definition:
Convenience sampling means selecting samples that are easy to reach or access, not necessarily random or representative.

ğŸ’¡ Why we use it:
Itâ€™s quick, cheap, and easy â€” used in early research or pilot studies.
But âš ï¸ high bias risk â€” results may not represent the true population.

ğŸ§  Example:
You want to know students' opinions about online classes â€”
Instead of surveying the whole college, you just ask your friends or classmates because theyâ€™re easy to reach.

Thatâ€™s convenience sampling.

"""

data = pd.DataFrame({
    'student_id': range(1,101),
    'feedback_score': np.random.randint(1,6,100)
})

# Convenience sample: just take first 10 entries
sample = data.head(10)

print(sample)
```

```python
""" 
Mean (Average)
The sum of all values divided by the total number of values.


ğŸ§® Example:
Suppose you have exam scores = [60, 70, 80, 90, 100]
Mean = (60 + 70 + 80 + 90 + 100) / 5 = 80

Mean is often used to fill missing data (imputation) or calculate average accuracy/loss over epochs in model training.
"""

scores = [60, 70, 80, 90, 100]
mean_score = np.mean(scores)
print(mean_score)  # 80.0
```

```python
""" 
Median â€” The Middle Value

Definition:
Median is the middle value in a sorted dataset, it divides the data into two equal halves.

ğŸ“˜ Formula / Steps:
â€” Arrange the data in ascending order.
â€” If the number of data points (n) is odd, median = middle value.
â€” If n is even, median = average of the two middle values.


ğŸ’¡ Use Case / Why Median is Better Than Mean Sometimes:
â€” Median is not affected by outliers (extremely high or low values).
â€” In skewed data (like income, house prices), the mean can be misleading.

Example:
Income = [20k, 25k, 30k, 35k, 10,00,000]
Mean = ~222,000 (heavily pulled by outlier)
Median = 30k (represents actual middle of most people)

ğŸ“Š In AI/ML:
â€” Used for robust imputation of missing values when data has outliers.
(e.g., replacing NaNs with median instead of mean)
â€” Useful in feature scaling and detecting skewness in distributions.
"""

data = [10, 20, 30, 40, 100]
median_value = np.median(data)
print(median_value)  # 30.0

```

```python
""" 
Mode â€” The Most Frequent Value

Mode is the value that appears most frequently in a dataset.
Unlike mean or median, mode is about frequency, not position or magnitude.


ğŸ’¡ Use Case / When Mode is Better:
â€” Works great with categorical data (like colors, cities, brands, etc.)
    Example â†’ ["Red", "Blue", "Red", "Green", "Red"] â†’ Mode = "Red"
â€” Useful when numbers donâ€™t make sense to average (e.g., zip codes, product categories)

In AI/ML:
â€” For categorical missing value imputation
â†’ replace NaN with most frequent category (mode)
â€” For understanding dominant class in classification datasets (like majority label)
"""

from statistics import mode

data = [2,3, 4, 4, 5, 4, 6, 7]
print(mode(data))  # 4

#If your data has multiple modes, you may need to handle that seperately
from scipy import stats
print(stats.mode(data, keepdims=True).mode[0])  # 4


# âœ… Alternatives for Mode

# You can use either Pythonâ€™s built-in or SciPy:

# ğŸ”¹ 1. Using statistics module (built-in, small datasets)
from statistics import mode
data = [1, 2, 2, 3, 3, 3, 4]
print(mode(data))  # 3

# ğŸ”¹ 2. Using scipy.stats.mode() (best for arrays)
from scipy import stats
import numpy as np

arr = np.array([1, 2, 2, 3, 3, 3, 4])
result = stats.mode(arr, keepdims=True)
print(result.mode[0])      # 3
print(result.count[0])     # how many times it appears


# ğŸ’¡ Pro tip:
# If youâ€™re doing data analysis or ML preprocessing, youâ€™ll mostly use pandas, which has a built-in:

import pandas as pd
s = pd.Series([1, 2, 2, 3, 3, 3, 4])
print(s.mode().iloc[0])  # 3
```
